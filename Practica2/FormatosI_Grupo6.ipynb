```python
import csv
csvarchivo = open('AccidentesBicicletas_2021.csv',encoding="utf8",errors='ignore')
entrada = csv.reader(csvarchivo, delimiter=";")
M = list(entrada)
```

```python
#Consulta 1

diccionario = dict()

for i in range(1, len(M)):
    tupla = (M[i][6], M[i][7])
    
    if tupla in diccionario:
        diccionario[tupla] += 1
    else:
        diccionario[tupla] = 1

for x in diccionario.items():
    print("En el distrito " + x[0][0] + " ha habido " + str(x[1]) + " accidentes de tipo " + x[0][1])
```

```python
#Consulta 2

diames = input("Introduzca el día y el mes (en formato DD/MM)")

fecha = diames + "/2021"


dLesion = dict()
dRango = dict()

for x in M:
    if (x[1] == fecha):
        lesion = x[14] #Revisar lesion-distrito
        distrito = x[6]
        if lesion in dLesion:
            if distrito in dLesion[lesion]:
                dLesion[lesion][distrito] += 1
            else:
                dLesion[lesion][distrito] = 1
        else:
            dLesion[lesion] = {distrito:1}
        
        rango = x[11] #Revisar rango-distrito
        if rango in dRango:
            if distrito in dRango[rango]:
                dRango[rango][distrito] += 1
            else:
                dRango[rango][distrito] = 1
        else:
            dRango[rango] = {distrito:1}
        
def obtener_valor(tupla):
     return tupla[1]

print("Para la fecha " + fecha + ":\n")
    
for x in dLesion.items():
    maximo = max(x[1].items(), key=obtener_valor)
    print("La lesión " + x[0] + " se ha repetido más veces (" + str(maximo[1]) + ") en el distrito " + maximo[0])

print()
    
for x in dRango.items():
    maximo = max(x[1].items(), key=obtener_valor)
    print("En el rango de edad " + x[0] + " se ha repetido más veces (" + str(maximo[1]) + ") en el distrito " + maximo[0])
    
```

```python
#Consulta 3
def menu1() -> str or None:

    distritos = sorted(list(set(fila[6] for fila in M[1:])))
    
    for i, distrito in enumerate(distritos):
        print(f"{i+1}: {distrito}")

    opcion = int(input("\nSeleccione un distrito o ninguno (0):"))
    print("\n")
    
    
    return distritos[opcion - 1] if opcion > 0 else None

def menu2() -> str or None:
    
    lesiones = sorted(list(set(fila[14] for fila in M[1:] if fila[14] != "NULL")))
    
    for i, lesion in enumerate(lesiones):
        print(f"{i+1}: {lesion}")

    opcion = int(input("\nSeleccione una lesión o ninguna(0):"))
    print("\n")

    return lesiones[opcion - 1] if opcion > 0 else None

def menu3() -> str or None:
    
    edades = sorted(list(set(fila[11] for fila in M[1:] if fila[11] != "Desconocido")))
    edades.insert(0, edades.pop(-2)) # Para poner "Menor de 5" al principio
    
    for i, edad in enumerate(edades):
        print(f"{i+1}: {edad}")

    opcion = int(input("\nSeleccione una rango de edad o ninguno(0):"))
    print("\n")
    
    return edades[opcion - 1] if opcion > 0 else None

distrito = menu1()
lesion = menu2()
edad = menu3()

res = M

if distrito:
    res = [x for x in res if x[6] == distrito]

if lesion:
    res = [x for x in res if x[14] == lesion]
    
if edad:
    res = [x for x in res if x[11] == edad]
    
calles = {}
for fila in res:
    
    try:
        calles[fila[3]] += 1
    
    except KeyError:
        calles[fila[3]] = 1

if len(calles) == 0:
    print("No hay accidentes que cumplan con esos criterios.")
    
else:
    
    top_calles = sorted(calles.items(), key=lambda x: x[1], reverse=True)
    for num, calle in enumerate(top_calles[:10]):
        print(f"{num+1}. {calle[0]}: {calle[1]} accidentes.")
```

```python
#Consulta 4

def menu4() -> str:
  opciones = {'tipo de accidente':'tipo_accidente' , 'rango de edad': 'rango_edad', 'tipo de lesividad':'tipo_lesividad'}
  claves = list(opciones.keys())
  for i, c in enumerate(claves):
    print(str(i + 1) + ':' + (c))
  print('\n')
  op = int(input("\nSeleccione una de las opciones:"))
  return opciones[claves[op - 1]]

def menu5(op) -> str or None:
  pos = M[0].index(op)
  valores = list(set(fila[pos] for fila in M[1:]))
  for i, v in enumerate(valores):
    print(str(i + 1) + ':' + str(v))
  print('\n')
  valor = int(input("\nSeleccione uno de los valores(0-ninguno):"))
  return valores[valor - 1] if valor > 0 else None
```

```python

def filtrar_calles_por_valor_campo(campo, valor):
  calles = dict()
  pos = M[0].index(campo) # Columna del campo en la cabecera
  for f in M[1:]:
    if(f[pos] == valor):#Conteo de numero de accidente
      try:
        calles[f[3]] += 1
      except KeyError:
        calles[f[3]] = 1
  return calles
  
def mostrar_calles(calles):
  for c, num_acc in calles.items():
    print(f"Calle:{c} . Numero total de accidentes:{num_acc}" )

```

```python
campo = menu4() #Campo
valor = menu5(campo) # Valor de campo

if valor:
  print('Calles relacionados con ' + campo + ': ' + valor)
  calles = filtrar_calles_por_valor_campo(campo, valor)
  mostrar_calles(calles)
else:
  pos = M[0].index(campo)
  
  todo_valor_campo = list(set(fila[pos] for fila in M[1:]))
  for v in todo_valor_campo:
    filtrado = filtrar_calles_por_valor_campo(campo, v)
    if len(filtrado) > 0:
      print('Calles relacionados con ' + campo + ': ' + str(v))
      mostrar_calles(filtrado)
    print('\n')
```

A continuación, debeis calcular la distancia entre cada monumento y el lugar indicado. Esto se puede hacer usando la siguiente función que os propociono. 

```python
from math import radians, cos, sin, asin, sqrt
def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    # Radius of earth in kilometers is 6371
    km = 6371* c
    return km
#Ejemplo de distancia entre la facultad y el Ministerio de Agricultura
haversine (-3.6906142029315534,40.408789366623914,  -3.7371139,40.4520948)
```
